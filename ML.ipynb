```python
import re
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import string
import nltk
import warnings 

from sklearn.metrics.pairwise import cosine_similarity
warnings.filterwarnings("ignore", category=DeprecationWarning)

%matplotlib inline
```

```python
data = pd.read_csv('android.csv', delimiter=",",engine='python', encoding='utf8')
```

```python
data.head(20)
```

```python
data.drop('Timestamp', axis=1, inplace=True)
```

```python
#input_text ->original text
#pattern -> pattern we want to remove
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt  
```

```python
# remove special characters, numbers, punctuations
data['cleaned_para'] = data['Lists your Projects/ Achievements'].str.replace("[^a-zA-Z]", " ")
```

```python
data.head()
```

```python
# remove short words
data['cleaned_para'] = data['cleaned_para'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
data.head()
```

```python
tokenized_data = data['cleaned_para'].apply(lambda x: x.split())
```

```python
tokenized_data.head()
```

```python
# stemming (play, players etc to play)
from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_data = tokenized_data.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
```

```python
tokenized_data.head()
```

```python
data.head()
```

```python
#replacing new with result
for i in range(len(tokenized_data)):
    tokenized_data[i] = ' '.join(tokenized_data[i])

data['cleaned_para'] = tokenized_data
```

```python
data.head()
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
# bag-of-words feature matrix
bow = bow_vectorizer.fit_transform(data['cleaned_para'])
```

```python
bow
```

```python
pd.DataFrame(bow.todense()).head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
# TF-IDF feature matrix
tfidf = tfidf_vectorizer.fit_transform(data['cleaned_para'])
```

```python
pd.DataFrame(tfidf.todense()).head(25)
```

```python
# generating the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf, tfidf)
```

```python
cosine_sim
```
